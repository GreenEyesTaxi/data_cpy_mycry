```python
import pandas as pd
```

```python
file = pd.read_csv('Documents/analisys/passengers.csv',low_memory=False)
```

```python
def check_age(row):
    birth_date_row = row['Birth Date']
    flight_date_row = row['Flight Date']
    try:
        birth_date = [int(birth_date_row[:2]), int(birth_date_row[3:5]), int(birth_date_row[6:10])]
        flight_date = [int(flight_date_row[5:7]), int(flight_date_row[8:10]), int(flight_date_row[:4])]
    except (TypeError, ValueError):
        return False
    check = False

    if flight_date[2] - birth_date[2] <= 18:
        if flight_date[2] - birth_date[2] == 18:
            if flight_date[0] < birth_date[0]:
                check = True
            elif flight_date[0] == birth_date[0]:
                if flight_date[1] < birth_date[1]:
                    check = True
        else:
            check = True

    return check
    
file['Under 18'] = file.apply(check_age, axis=1)
```

```python
under_check_all = file[(file['Last Name'].notna()) & (file['FlightTime'].notna()) & (file['Flight Date'].notna())][['Last Name', 'Under 18', 'Flight Date', 'FlightTime']].copy()
under_check_childrens = under_check_all[under_check_all['Under 18'] == True][['Last Name','Flight Date','FlightTime']].copy()
under_check_parents = under_check_all[under_check_all['Under 18'] == False][['Last Name','Flight Date','FlightTime']].copy()
```

```python
under_check_childrens['Index'] = under_check_childrens.index
merged_df = pd.merge(
    under_check_childrens,
    under_check_parents,
    on=['Last Name', 'Flight Date', 'FlightTime'],
    how='inner'
)
merged_df.set_index('Index', inplace=True)
```
